## Problem 6

Create two scatter plots that shows the PRICE in the $y$ axis and the SQUARE FEET in the $x$-axis.

1) One scatter plot should be in a normal scale.

2) The other scatter plot should have a log-log scale.

```python
### Write the code here
# PRICE in the y-axis.
# SQUARE FEET in the x-axis.



plt.figure(figsize=[14,6])
x=small_data['SQUARE FEET']
y=small_data['PRICE']



# 1

plt.subplot(1,2,1)
plt.scatter(x,y,c='#3CB371',alpha = 0.4,s=40)

# 2

plt.subplot(1,2,2)
plt.loglog(x,y,'.',alpha = 0.9,c='#98FB98');




```
